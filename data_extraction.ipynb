Let's start segregation


```python
grp_list = {'CN' : [], 
            'EMCI' : [],
            'LMCI' : [],
            'MCI' : [],
            'AD' : [],
            'SMC' : [],
            'Patient' : []}
```

```python

root_dir = 'G:\ADNI\ADNI'
contents = os.listdir(root_dir)
new_dir = 'G:\ADNI\Segregated'
test_contents = contents
# print(len(test_contents))
```

```python
for grp in grp_list:
    grp_list[grp].clear()
```

```python
def traverse_folder(path):
    d = os.listdir(path)
    last_folder = None
    for item in d:
        item_path = os.path.join(path, item)
        if os.path.isdir(item_path):
            last_folder = traverse_folder(item_path)
        else:
            folders = os.path.normpath(path).split(os.path.sep)
            last_folder = folders[-1]
    return last_folder
```

```python
for item in test_contents:
    subject_id = item
    item_path = os.path.join(root_dir,item)
    # print(item_path)
    fmri_id = traverse_folder(item_path)
    # print(fmri_id)
    for index, row in df.iterrows():
        subject = row['Subject']
        img_id = row['Image Data ID']
        group = row['Group']
        if subject == subject_id and img_id == fmri_id:
            if group in grp_list:
                    grp_list[group].append(subject_id)
```

```python
for group, subjects in grp_list.items():
    print(f"{group}: {subjects}")
```

```python
for item in contents:
    item_path = os.path.join(root_dir,item) #sorce dir
    for catagory, subject_ids in grp_list.items():
        new_path = os.path.join(new_dir,catagory)
        if item in subject_ids:
            shutil.move(item_path,new_path)
```
